```python
!pip install scikit-optimize
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn import metrics
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV, cross_val_score
from skopt import BayesSearchCV
from skopt.space import Integer, Categorical
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# read the data and create database

train_df=pd.read_csv('/content/drive/MyDrive/machine_learning_dataset/Insurance Fraud - TRAIN-3000.csv')
train_df.info() , train_df.head()
```

```python
# Check for null values

train_df.isnull().sum()
```

```python
# define Categorical coulmn and numerical columns

categorical_col=train_df.select_dtypes(include=['object']).columns
numerical_col=train_df.select_dtypes(exclude=['object']).columns

print(categorical_col)
print(numerical_col)
```

```python
# label encodingthe categorical columns

le=LabelEncoder()
for col in categorical_col:
  train_df[col]=le.fit_transform(train_df[col])
```

```python
# chcek if data set is balanced or not

train_df['FRAUDFOUND'].value_counts()
```

```python
# Define featureas and traget

X=train_df.drop(columns=['FRAUDFOUND'])
y=train_df['FRAUDFOUND']

# apply smote to balalnce it
smote= SMOTE()

X_resampled,y_resampled=smote.fit_resample(X,y)

X_resampled.value_counts()
y_resampled.value_counts()
```

```python
# split the datatset into the train and test

X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

# Decison Tree Classifier

dt_model=DecisionTreeClassifier()
dt_model.fit(X_train,y_train)
```

```python
# random forest Classifier

rf_model=RandomForestClassifier()
rf_model.fit(X_train,y_train)
```

```python
# read the test data

test_df=pd.read_csv('/content/drive/MyDrive/machine_learning_dataset/Insurance Fraud -TEST-12900.csv')

# categorical columns and numerical cols

categorical_col1=test_df.select_dtypes(include=['object']).columns
numerical_col2=test_df.select_dtypes(exclude=['object']).columns

# encode the data set

le=LabelEncoder()
for col in categorical_col1:
  le.fit(test_df[col])
  test_df[col]=le.transform(test_df[col])

y_test_actual=test_df['FRAUDFOUND']
test_df=test_df.drop(columns=['FRAUDFOUND'])

# smote= SMOTE()

# test_df,y_test=smote.fit_resample(X,y)

X_test.shape


```

```python
# Predict the data set
pred_train_dt=dt_model.predict(X_test)
pred_test_dt=dt_model.predict(test_df)
```

```python
#Model Accuracy
print("Train Accuracy:", metrics.accuracy_score(y_test,pred_train_dt))
print("Test Accuracy:", metrics.accuracy_score(y_test_actual,pred_test_dt))
print("Confusion Matrix for Decision Tree:")
print(confusion_matrix(y_test_actual,pred_test_dt))
print("Max Depth",dt_model.get_depth())
print("Leaf",dt_model.get_n_leaves())
print('Printing the precision and recall, among other metrics')
print(metrics.classification_report(y_test_actual, pred_test_dt))
```

```python
pred_train_rf=rf_model.predict(X_test)
pred_test_rf=rf_model.predict(test_df)
#Model Accuracy
print("Train Accuracy:", metrics.accuracy_score(y_test,pred_train_rf))
print("Test Accuracy:", metrics.accuracy_score(y_test_actual,pred_test_rf))
print("Confusion Matrix for Decision Tree:")
print(confusion_matrix(y_test_actual,pred_test_rf))
print('Printing the precision and recall, among other metrics')
print(metrics.classification_report(y_test_actual,pred_test_rf))
```

```python
#Hyperparametr tuning for Decision Tree Classifier

# RANDOM SEARCH ---------

print("RandomizedSearchCV-Decision tree")
parameters={'min_samples_leaf' : range(10,300,10),'max_depth': range(5,30,2),'criterion':['gini','entropy']}
dt_random = RandomizedSearchCV(dt_model,parameters,n_iter=25,cv=5)
dt_random.fit(X_train, y_train)
grid_parm=dt_random.best_params_
print(grid_parm)
print("accuracy Score for Decision Tree:{0:6f}".
      format(dt_random.score(test_df, y_test_actual)))

# Chnanging hyperparameter variable

print("\nRandomizedSearchCV-Decision tree")

# Expanded hyperparameter ranges and added 'min_samples_split'
parameters = {
    'min_samples_leaf': range(1, 50, 5),
    'max_depth': range(1, 20, 2),
    'criterion': ['gini', 'entropy'],
    'min_samples_split': range(2, 10, 2),
    'max_features': ['sqrt', 'log2', None]}

dt_random = RandomizedSearchCV(dt_model, parameters, n_iter=50, cv=5) # Changed: Increased n_iter
dt_random.fit(X_train, y_train)

grid_parm = dt_random.best_params_
print(grid_parm)

print("accuracy Score for Decision Tree: {0:6f}".format(dt_random.score(test_df, y_test_actual)))
```

Explaination

The second tuning round yielded better results due to finer adjustments in min_samples_split and max_features. These additional parameters helped improve model performance, as seen in the increased accuracy score.

```python
#Hyperparameter tuning done for random forest classifier

#RANDOM SEARCH--------------------------------------------

print("RandomizedSearchCV-Random forest")
rand_parameters={'min_samples_leaf' : range(10,100,10),'max_depth':
            range(5,30,2),'max_features':[5,30,2],'n_estimators':[20,30,40]}
rf_random = RandomizedSearchCV(rf_model,rand_parameters,n_iter=25,cv=5)
rf_random.fit(X_train, y_train)
grid_parm=rf_random.best_params_
print(grid_parm)
print("accuracy Score for Decision Tree:{0:6f}".
      format(rf_random.score(test_df,y_test_actual)))

# Chnanging hyperparameter variable

print("\nRandomizedSearchCV-Random forest")
rand_parameters={
    'min_samples_leaf' : range(20,200,20),
    'max_depth': range(10,40,8),
    'max_features':[5,30,2],
    'n_estimators':[20,30,40]
}
rf_random = RandomizedSearchCV(rf_model,rand_parameters,n_iter=50,cv=5)
rf_random.fit(X_train, y_train)
grid_parm=rf_random.best_params_
print(grid_parm)
print("accuracy Score for Decision Tree:{0:6f}".
      format(rf_random.score(test_df,y_test_actual)))
```

# ***Cross Validation for each type of Classifier***

```python
# Decision Tree

# Define parameter grid

parameters_dt= {
    'min_samples_leaf': range(1, 50, 5),
    'max_depth': range(1, 20, 2),
    'criterion': ['gini', 'entropy'],
    'min_samples_split': range(2, 10, 2),
    'max_features': ['sqrt', 'log2', None]
}

parameters_bayes = {
    'min_samples_leaf': Integer(1, 50),
    'max_depth': Integer(1, 20),
    'criterion': Categorical(['gini', 'entropy']),
    'min_samples_split': Integer(2, 10),
    'max_features': Categorical(['sqrt', 'log2', None])
}

# RANDOM SEARCH with Cross-Validation
print("RandomizedSearchCV - Decision Tree")
dt_random = RandomizedSearchCV(dt_model, parameters_dt, n_iter=50, cv=5, random_state=42)
dt_random.fit(X_train, y_train)
random_best_params = dt_random.best_params_
random_best_cv_score = cross_val_score(dt_random.best_estimator_, X_train, y_train, cv=5).mean()
random_test_score = accuracy_score(y_test_actual, dt_random.predict(test_df))
print("Best Parameters (Random Search):", random_best_params)
print("Cross-Validated Accuracy (Random Search):", random_best_cv_score)
print("Test Accuracy (Random Search):", random_test_score)

# GRID SEARCH with Cross-Validation
print("\nGridSearchCV - Decision Tree")
dt_grid = GridSearchCV(dt_model, parameters_dt, cv=5)
dt_grid.fit(X_train, y_train)
grid_best_params = dt_grid.best_params_
grid_best_cv_score = cross_val_score(dt_grid.best_estimator_, X_train, y_train, cv=5).mean()
grid_test_score = accuracy_score(y_test_actual, dt_grid.predict(test_df))
print("Best Parameters (Grid Search):", grid_best_params)
print("Cross-Validated Accuracy (Grid Search):", grid_best_cv_score)
print("Test Accuracy (Grid Search):", grid_test_score)

# BAYESIAN SEARCH with Cross-Validation
print("\nBayesSearchCV - Decision Tree")
dt_bayes = BayesSearchCV(dt_model, parameters_bayes, n_iter=50, cv=5, random_state=42)
dt_bayes.fit(X_train, y_train)
bayes_best_params = dt_bayes.best_params_
bayes_best_cv_score = cross_val_score(dt_bayes.best_estimator_, X_train, y_train, cv=5).mean()
bayes_test_score = accuracy_score(y_test_actual, dt_bayes.predict(test_df))
print("Best Parameters (Bayesian Search):", bayes_best_params)
print("Cross-Validated Accuracy (Bayesian Search):", bayes_best_cv_score)
print("Test Accuracy (Bayesian Search):", bayes_test_score)

```

```python
# Cross validation for Random forest
#Define Parameter grid

parameters_rf = {
    'n_estimators': [50, 300, 50],
    'min_samples_leaf': range(1, 50, 5),
    'max_depth': range(1, 20, 2),
    'criterion': ['gini', 'entropy'],
    'min_samples_split': range(2, 10, 2),
    'max_features': ['sqrt', 'log2', None]
}

parameters_bayes1 ={
    'n_estimators': Integer(50, 300),
    'max_depth': Integer(5, 30),
    'min_samples_split': Integer(2, 10),
    'min_samples_leaf': Integer(1, 10),
    'max_features': Categorical(['sqrt', 'log2', None]),
    'criterion': Categorical(['gini', 'entropy'])
}

# RANDOM SEARCH FOR CROSS VALIDATIO
print("RandomizedSearchCV - Random Forest")
rf_random=RandomizedSearchCV(rf_model, parameters_rf, n_iter=50, cv=5, random_state=42)
rf_random.fit(X_train, y_train)
rf_random_best_params= rf_random.best_params_
rf_random_best_cv_score=cross_val_score(rf_random.best_estimator_, X_train, y_train, cv=5).mean()
rf_random_test_score=accuracy_score(y_test_actual, rf_random.predict(test_df))
print("Best Parameters (Random Search):", random_best_params)
print("Cross-Validated Accuracy (Random Search):", random_best_cv_score)
print("Test Accuracy (Random Search):", random_test_score)

# GRID SEARCHH FOR CROSS VALIDATION
# print("\nGridSearchCV - Random Forest")
# rf_grid=GridSearchCV(rf_model, parameters_rf, cv=3, n_jobs=-1)
# rf_grid.fit(X_train, y_train)
# rf_grid_best_params=rf_grid.best_params_
# rf_grid_best_cv_score=cross_val_score(rf_grid.best_estimator_, X_train, y_train, cv=3).mean()
# rf_grid_test_score=accuracy_score(y_test_actual, rf_grid.predict((test_df)))
# print("Best Parameters (Grid Search):", grid_best_params)
# print("Cross-Validated Accuracy (Grid Search):", grid_best_cv_score)
# print("Test Accuracy (Grid Search):", grid_test_score)

# BAYESIAN SEARCH with Cross-Validation'
print("\nBayesSearchCV - Random Forest")
rf_bayes=BayesSearchCV(rf_model, parameters_bayes1, n_iter=50, cv=5, random_state=42)
rf_bayes.fit(X_train, y_train)
rf_bayes_best_params=rf_bayes.best_params_
rf_bayes_best_cv_score=cross_val_score(rf_bayes.best_estimator_, X_train, y_train, cv=5).mean()
rf_bayes_test_score=accuracy_score(y_test_actual, rf_bayes.predict(test_df))
print("Best Parameters (Bayesian Search):", bayes_best_params)
print("Cross-Validated Accuracy (Bayesian Search):", bayes_best_cv_score)
print("Test Accuracy (Bayesian Search):", bayes_test_score)
```

```python
# Build decision treee on best parameter

random_best_model=DecisionTreeClassifier(**random_best_params, random_state=42)
grid_best_model=DecisionTreeClassifier(**grid_best_params, random_state=42)
bayes_best_model=DecisionTreeClassifier(**bayes_best_params, random_state=42)

#fit model on training data

random_best_model.fit(X_train,y_train)
grid_best_model.fit(X_train,y_train)
bayes_best_model.fit(X_train,y_train)

# Evaluate the model

models_dt = {
    "Decisom Tree (random search)": random_best_model,
    "Decisom Tree (grid search)": grid_best_model,
    "Decisom Tree (bayes search)": bayes_best_model
}

for model_name, model in models_dt.items():
    y_pred=model.predict(test_df)
    print(f"\n{model_name}:")
    print("Accuracy:", accuracy_score(y_test_actual, y_pred))
    print("Precision:", precision_score(y_test_actual, y_pred))
    print("Recall:", recall_score(y_test_actual, y_pred))
    print("F1 Score:", f1_score(y_test_actual, y_pred))
```
